# Concatenating Arrays

```python
import numpy as np
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
```

```python
res = np.concatenate([arr1,arr2])
res
```

```python
arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])
arr2 = np.array([[7, 8, 9],
                 [10, 11, 12]])
```

```python
res = np.concatenate([arr1, arr2], axis=1)
res
```
