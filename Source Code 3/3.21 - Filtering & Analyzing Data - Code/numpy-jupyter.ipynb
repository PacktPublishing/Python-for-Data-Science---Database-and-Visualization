### Data Exploration with NumPy

```python
import numpy as np
```

### Sales Data, 7 Days, 3 Products

```python
sales_data = np.array([
    [120, 95, 200],
    [130, 100, 210],
    [125, 90, 190],
    [140, 110, 220],
    [135, 105, 215],
    [145, 115, 230],
    [150, 120, 240]
])
sales_data
```

```python
sales_data.shape
```

#### Total Sales Per Product

```python
total_per_product = sales_data.sum(axis=0)
total_per_product
```

#### Total Sales Per Day

```python
total_per_day = sales_data.sum(axis=1)
total_per_day
```

```python
max_day_sales = total_per_day.max()
print(max_day_sales)
```

```python
best_day_index = total_per_day.argmax()
print(best_day_index)
```

#### Average Sales Per Product

```python
avg_per_product = sales_data.mean(axis=0)
print(avg_per_product)
```

```python
best_product_index = total_per_product.argmax()
print(best_product_index)
```

#### Sales Values Greater Than 130

```python
high_sales = sales_data[sales_data > 130]
high_sales
```

```python
count_high_sales = (sales_data > 130).sum()
print(count_high_sales)
```

```python
total_per_day = sales_data.sum(axis=1)
total_per_day
```

#### Days with Total Sales Above 400

```python
good_days = total_per_day > 400
good_days
```

```python
total_per_day[good_days]
```

```python
print(good_days.sum())
```

#### Product 1 Sales Analysis

```python
product_1_sales = sales_data[:,0]
product_1_sales
```

```python
product_1_avg = product_1_sales.mean()
print(product_1_avg)
```

```python
above_avg_days = product_1_sales > product_1_avg
above_avg_days
```

```python
product_1_sales[above_avg_days]
```

```python
percentage = (above_avg_days.sum()/ len(product_1_sales)) * 100
print(percentage)
```
