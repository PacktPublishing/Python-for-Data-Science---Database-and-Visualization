# Indexing & Slicing

```python
import numpy as np
```

```python
arr = np.array([10,20,30,40,50])
arr
```

```python
print(arr[0])
```

```python
print(arr[2])
```

```python
print(arr[-1])
```

```python
print(arr[-2])
```

```python
arr[1:4]
```

```python
arr[:3]
```

```python
arr[2:]
```

```python
arr[::2]
```

```python
arr[::-1]
```

```python
arr[0] = 100
arr
```

```python
arr[1:3] = [200, 300]
arr
```

```python
slice_arr = arr[3:5]
slice_arr
```

```python
slice_arr[0] = 400
arr
```

```python
slice_arr = arr[3:5].copy()
slice_arr[1] = 500
```

```python
arr
```

```python
slice_arr
```
