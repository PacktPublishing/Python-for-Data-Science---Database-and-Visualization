# Masking Arrays

```python
import numpy as np
```

```python
arr = np.array([10,15,20,25,30,35,40])
arr
```

```python
mask = arr > 20
mask
```

```python
res = arr[mask]
res
```

```python
even_mask = arr % 2 == 0
print(even_mask)
print(arr[even_mask])
```

```python
mask = arr > 25
arr[mask] = 0
arr
```
