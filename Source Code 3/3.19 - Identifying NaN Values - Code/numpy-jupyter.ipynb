# Dealing with Missing Values

```python
import numpy as np
```

```python
data = [10, 20, None, 40, '', 60, None, 80]
data
```

```python
arr = np.array(data)
arr
```

```python
cleaned_data = [np.nan if(x is None or x == '') else x for x in data]
cleaned_data
```

```python
arr = np.array(cleaned_data, dtype=float)
arr
```

```python
arr == np.nan
```

```python
missing_val = np.isnan(arr)
missing_val
```
