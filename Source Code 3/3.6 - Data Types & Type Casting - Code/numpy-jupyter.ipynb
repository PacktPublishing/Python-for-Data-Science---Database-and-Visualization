# Data Types & Type Casting

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4.2,5])
arr.dtype
```

```python
lst = ["test", 1, 2, 3, 4]
arr = np.array(lst)
arr
```

```python
arr = np.array([1.7, 2.3, 3.9, 4.1])
arr
```

```python
arr_int = arr.astype(int)
arr_int
```

```python
arr_str = arr_int.astype(str)
arr_str
```

```python
arr = np.array([1, 2, 3, 4], dtype='uint8')
arr.dtype
```

```python
arr = np.array(["1", "2", "test"])
arr2 = arr.astype(np.int64)
arr2
```
