# Conditional Based Operations - Pt 2

```python
import numpy as np
```

```python
arr = np.array([10, 25, 30, 45, 50, 15])
arr
```

```python
condition = arr > 20
condition
```

```python
result = arr[(arr < 20) | (arr > 40)]
result
```

```python
res = np.where(arr < 20, "low", "high")
res
```

```python
print(np.argwhere(arr > 25))
```

```python
arr_2d = np.array([[1,2,3], 
                   [4,5,6]])
np.argwhere(arr_2d > 2)
```
