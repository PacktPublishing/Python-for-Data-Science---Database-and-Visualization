```python
import matplotlib.pyplot as plt
```

```python
scores = [45, 52, 58, 61, 63, 65, 67, 68, 70, 71,
          72, 73, 74, 75, 78, 80, 83, 85, 88, 92]

plt.hist(scores, bins=10, color="steelblue", edgecolor="black")
plt.title("Student Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.grid(axis="y")
plt.show()
```
