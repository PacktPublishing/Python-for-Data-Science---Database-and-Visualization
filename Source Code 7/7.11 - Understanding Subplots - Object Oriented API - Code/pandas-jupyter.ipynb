```python
import matplotlib.pyplot as plt
```

```python
months  = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [12000, 15500, 13200, 17800, 16400, 20100]
expense = [8000,  9200,  8700, 10500,  9800, 11200]
profit  = [4000,   6300,  4500,  7300,  6600,  8900]
scores  = [45, 52, 58, 61, 65, 68, 70, 72, 74, 75,
           78, 80, 80, 83, 85, 85, 88, 88, 90, 92]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

axes[0][0].plot(months, revenue, color="steelblue", linewidth=2, marker="o")
axes[0][0].set_title("Revenue")
axes[0][0].set_xlabel("Month")
axes[0][0].set_ylabel("Amount ($)")
axes[0][0].grid(True)

axes[0][1].plot(months, expense, color="tomato", linewidth=2, marker="s")
axes[0][1].set_title("Expenses")
axes[0][1].set_xlabel("Month")
axes[0][1].set_ylabel("Amount ($)")
axes[0][1].grid(True)

axes[1][0].bar(months, profit, color="mediumseagreen", edgecolor="black")
axes[1][0].set_title("Monthly Profit")
axes[1][0].set_xlabel("Month")
axes[1][0].set_ylabel("Profit ($)")
axes[1][0].grid(axis="y")

axes[1][1].hist(scores, bins=8, color="orchid", edgecolor="black")
axes[1][1].set_title("Score Distribution")
axes[1][1].set_xlabel("Score")
axes[1][1].set_ylabel("Count")
axes[1][1].grid(axis="y")
plt.tight_layout()
plt.show()

# print(type(axes))
# print(axes.shape)
```
