```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5] 
y = [10, 20, 30, 40, 50]
z = [100, 200, 300, 400, 500]

plt.plot(x, y, color="green", linestyle="--", linewidth=2, marker="s", label="Line A")
plt.plot(x, z, color="orange", linestyle="-", linewidth=2, marker="o", label="Line B")
plt.title("My First Plot") 
plt.xlabel("X Axis") 
plt.ylabel("Y Axis") 
plt.legend()
plt.grid()
plt.show()
```
