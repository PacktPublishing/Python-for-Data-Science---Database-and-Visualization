```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,5), dpi=300)

months  = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [12000, 15500, 13200, 17800, 16400, 20100]

plt.plot(months, revenue, color="steelblue", linewidth=2, marker="o")
plt.title("Monthly Revenue")
plt.ylabel("Revenue ($)")
plt.grid(True)
plt.show()
```
