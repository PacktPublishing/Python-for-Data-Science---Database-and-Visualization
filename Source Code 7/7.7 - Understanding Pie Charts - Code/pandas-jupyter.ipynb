```python
import matplotlib.pyplot as plt
```

```python
categories = ["Rent", "Food", "Transport", "Utilities", "Entertainment"]
expenses   = [12000, 5000, 2000, 1500, 2500]
colors = ["tomato", "steelblue", "orange", "mediumseagreen", "mediumpurple"]

plt.pie(expenses, labels=categories, autopct="%1.1f%%", 
        colors=colors, explode=[0.1,0,0,0,0])
plt.axis("equal")
plt.show()
```
