```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
products = ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard"]
sales_a = [320, 580, 210, 145, 430]
sales_b = [410, 620, 275, 190, 390]

x = np.arange(len(products))
width = 0.35

plt.bar(x - width / 2, sales_a, width=width, color="steelblue", label="Sale A")
plt.bar(x + width / 2, sales_b, width=width, color="tomato", label="Sale B")

plt.xticks(ticks=x, labels=products)
plt.title("A vs B Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales (in $)")
plt.legend()
plt.show()
```
