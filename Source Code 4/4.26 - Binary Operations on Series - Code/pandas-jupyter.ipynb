```python
import pandas as pd
```

```python
price = pd.Series(
    [1500, 800, 3500, 12000, 450],
    index=["Keyboard", "Mouse", "Tablet", "Laptop", "USB Hub"]
)
discount = pd.Series(
    [100, 50, 300, 1000, 30],
    index=["Keyboard", "Mouse", "Tablet", "Laptop", "USB Hub"]
)
```

```python
final_price = price - discount
final_price
```

```python
final_price = price + discount
final_price
```

```python
price * 2
```

```python
price / 2
```

```python
discount2 = pd.Series(
    [100, 50, 300, 1000, 200],
    index=["Keyboard", "Mouse", "Tablet", "Laptop", "Monitor"]
)
price - discount2
```

```python
price.sub(discount2, fill_value=0)
```
