```python
import pandas as pd
```

```python
orders = pd.DataFrame({
    "OrderID":    ["O_001", "O_002", "O_003", "O_004", "O_005", "O_006", "O_007"],
    "CustomerID": ["C01", "C02", "C01", "C03", "C02", "C04", "C99"],
    "Product":    ["Laptop Pro", "Wireless Mouse", "4K Monitor",
                   "HD Webcam", "Gaming Laptop", "USB Keyboard", "Random Item"],
    "Amount":     [45000, 800, 18000, 2500, 47000, 1400, 1200]
})

customers = pd.DataFrame({
    "CustomerID": ["C01", "C02", "C03", "C04"],
    "Name":       ["Cust_A", "Cust_B", "Cust_C", "Cust_D"],
    "City":       ["Florida", "Alaska", "Ohio", "New Jersey"]
})
```

```python
result = pd.merge(orders, customers, on="CustomerID")
print(result)
```

```python
result = pd.merge(orders, customers, on="CustomerID", how="left")
print(result)
```

```python
result = pd.merge(orders, customers, on="CustomerID", how="outer")
print(result)
```
