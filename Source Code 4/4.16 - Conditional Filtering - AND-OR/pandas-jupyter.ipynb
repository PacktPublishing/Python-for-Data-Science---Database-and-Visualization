```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
   "OrderID":  ["O_001", "O_002", "O_003", "O_004", "O_005",
                "O_006", "O_007", "O_008", "O_009", "O_010"],
   "Customer": ["Cust_A", "Cust_B", "Cust_C", "Cust_D", "Cust_E",
                "Cust_F", "Cust_G", "Cust_H", "Cust_I", "Cust_J"],
   "Product":  ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam",
                "Headset", "Laptop", "Mouse", "Monitor", "Keyboard"],
   "Amount":   [45000, 800, 1500, 12000, 2500,
                3500, 47000, 750, 11500, 1400],
   "Status":   ["Delivered", "Pending", "Delivered", "Shipped", "Delivered",
                "Pending", "Delivered", "Shipped", "Delivered", "Pending"]
})
df
```

```python
res = df.loc[(df["Status"] == "Delivered") & (df["Amount"] <= 10000)]
display(res)
print(f"Total rows: {len(res)}")
```

```python
res = np.where(df["Status"] == "Pending")[0]
display(df.loc[res,["Customer", "Product", "Status"]])
```

```python
res = df.loc[(df["Status"] == "Shipped") | (df["Amount"] < 1000)]
display(res)
```
