```python
df = df.drop_duplicates()
df.shape
```

```python
df = df.drop(columns=["show_id"])
df.shape
```

```python
df = df.rename(columns={
    "listed_in": "genre",
    "date_added": "date_added_on_netflix"
})
df.columns
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df
```

```python
df.columns = df.columns.str.upper()
df.columns
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.isnull().sum() / len(df) * 100
```

## Handling - director column

```python
df["DIRECTOR"] = df["DIRECTOR"].fillna("Unknown")
df["DIRECTOR"].isnull().sum()
```

## Handling - country column

```python
df["COUNTRY"].mode()[0]
```

```python
df["COUNTRY"] = df["COUNTRY"].fillna(df["COUNTRY"].mode()[0])
df["COUNTRY"].isnull().sum()
```

## Handling - date_added, rating and duration

```python
df = df.dropna(subset=["DATE_ADDED_ON_NETFLIX", "RATING", "DURATION"])
df.isnull().sum()
```

## Handling - cast column

```python
df["CAST"] = df["CAST"].fillna("Unknown")
df.isnull().sum()
```

```python
df
```

```python
df.info()
```

## Clean the date_added_on_netflix column

```python
df["DATE_ADDED_ON_NETFLIX"] = df["DATE_ADDED_ON_NETFLIX"].str.strip()
df["DATE_ADDED_ON_NETFLIX"] = pd.to_datetime(df["DATE_ADDED_ON_NETFLIX"])
df["DATE_ADDED_ON_NETFLIX"].dtype
```

```python
df.info()
```

```python
df["YEAR_ADDED"] = df["DATE_ADDED_ON_NETFLIX"].dt.year
df["MONTH_ADDED"] = df["DATE_ADDED_ON_NETFLIX"].dt.month
df[["DATE_ADDED_ON_NETFLIX", "YEAR_ADDED", "MONTH_ADDED"]].head()
```

## Clean the rating column

```python
df["RATING"].unique()
```

```python
df["RATING"] = df["RATING"].replace(r"\d+\s*min", pd.NA, regex=True)
```

```python
df["RATING"] = df["RATING"].fillna(df["RATING"].mode()[0])
df["RATING"].unique()
```
